```python
# 1 Задание
log = pd.read_csv('visit_log.csv', sep= ';')
log.head()
```

```python
def filtered(row):
    if row['traffic_source'] == 'yandex':
        return 'organic'
    elif row['traffic_source'] == 'google':
        return 'organic'
    elif row['traffic_source'] == 'paid':
        return 'ad'
    elif row['traffic_source'] == 'email':
        return 'ad'
    elif row['traffic_source'] == 'email':
        if row['region'] == 'Russia':
            return 'ad'
        else:
            return 'other'
    

log['sourse_type'] = log.apply(filtered, axis = 1)
log.head()
```

```python
# 3 Задание
data = pd.read_csv('ratings.csv')
data.head()
```

```python
raiting_count = data.groupby('userId').size().reset_index(name = 'count')
filtered_users = raiting_count [raiting_count['count'] > 100]
user_lifetime = data[data['userId'].isin(filtered_users['userId'])].groupby('userId').agg(
    min_timestamp = ('timestamp', 'min'),
    max_timestamp = ('timestamp', 'max')
).reset_index()
user_lifetime['lifetime'] = user_lifetime['max_timestamp'] - user_lifetime['min_timestamp']
print(user_lifetime)
```
